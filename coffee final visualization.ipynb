```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Coffe_sales.csv")
```

```python
df.head(5)
```

```python
df.groupby("coffee_name")["money"].sum()
```

```python
# 📊 Sales by Coffee Type (e.g., top-selling drinks)
```

```python
df.groupby("coffee_name")["money"].sum().plot(kind="bar",color=["red","green","blue"])
plt.title("coffee sales revenue by coffee name")
plt.xlabel("coffee name")
plt.ylabel("revenue")
print("The top selling coffee is latte that total revenue is 26875.30")
plt.savefig("1.png")
```

```python
# ⏰ Sales by Hour of Day (peak business hours)
```

```python
df.groupby("hour_of_day")["money"].sum().plot(kind="line")
plt.xlabel("hour_of_day")
plt.ylabel("money")
plt.title("sales by hour of a day")
print("peak hour of sales is 9 to 10 hour")
plt.savefig("2.png")
```

```python
df.groupby("hour_of_day")["money"].sum().plot(kind="bar")
plt.xlabel("hour_of_day")
plt.ylabel("money")
plt.title("sales by hour of a day")
print("peak hour of sales is 9 to 10 hour")
plt.savefig("3.png")
```

```python
# 🌅 Sales by Time of Day (Morning vs Afternoon vs Night trends)
```

```python
df.head()
```

```python
df.groupby("Time_of_Day")['money'].sum().plot(kind="bar")
plt.xlabel("time_of_day")
plt.ylabel("money")
plt.title("sales by time of a day")
print("hightest salling at night")
plt.savefig("4.png")
```

```python
df.groupby("Time_of_Day")['money'].sum().plot(kind="pie",autopct="%.2f")
plt.xlabel("time_of_day")
plt.ylabel("money")
plt.title("sales by time of a day")
plt.savefig("5.png")
```

```python
# 💳 Payment Method Breakdown (cash vs card usage)
df["cash_type"].value_counts().plot(kind="pie",autopct="%.2f%%")
plt.savefig("6.png")
```
