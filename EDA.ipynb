```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.read_csv("tips - tips.csv")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
tips.head(3)
```

```python
# bivariate Analysis
```

```python
sns.scatterplot(x=tips["total_bill"],        #df["total_bill"]
               y=tips["tip"])
```

```python
# hue, style,color,size ---> comprision of two columns
```

```python
# Multivariate Analysis
```

```python
sns.scatterplot(x="total_bill",y="tip",data=tips , hue=tips["sex"])
```

```python
sns.scatterplot(x="total_bill",
               y="tip",data=tips,
               hue=tips["sex"],style=tips["smoker"])
```

```python
# categorical-categorical Reation --->use crosstab
```

```python
#HeatMap(Categorical-Categorical)
p=pd.crosstab(tips["day"],tips["time"])
p
```

```python
sns.heatmap(p)
```
